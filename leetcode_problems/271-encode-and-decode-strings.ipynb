# Encode And Decode Strings

### Problem: Encode And Decode Strings
Design an algorithm to encode a list of strings to a string. The encoded string is then sent over the network and is decoded back to the original list of strings.

Implement the `Codec` class:
- `Codec()` Initializes the `Codec` object.
- `String encode(List<String> strs)` Encodes a list of strings to a single string.

- `List<String> decode(String s)` Decodes a single string to a list of strings.

### Example 1:

**Input:** strs = ["Hello","World"]

**Output:** ["Hello","World"]

### Example 2:

**Input:** strs = ["","Hello","World"]

**Output:** ["","Hello","World"]

### Constraints:

- `1 <= strs.length <= 200`
- `0 <= strs[i].length <= 200`
- `strs[i]` consists of any possible characters out of all the 256 ASCII characters.


```python
class Solution:
    def encode(self, strs: list[str]):
        if not len(strs):
            return []
            
        encoded_str = ""
        for i in strs:
            encoded_str += i + "#"
        return encoded_str

    def decode(self, s: str):
        decoded_str = []
        string = ""
        for i in s:
            if i == "#":
                decoded_str.append(string)
                string = ""
            else:
                string += i
        return decoded_str


soln = Solution()
encoded_str = soln.encode(["", "Hello", "world"])
print(encoded_str)
decoded_str = soln.decode(encoded_str)
print(decoded_str)
```
