Given an integer array `nums`, return an array answer such that `answer[i]` is equal to the product of all the elements of `nums` except `nums[i]`.

The product of any prefix or suffix of `nums` is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in `O(n)` time and without using the division operation.

 

### Example 1:

**Input:** nums = [1,2,3,4]

**Output:** [24,12,8,6]

### Example 2:

**Input:** nums = [-1,1,0,-3,3]

**Output:** [0,0,9,0,0]
 

### Constraints:

```2 <= nums.length <= 10^5```

```-30 <= nums[i] <= 30```

The input is generated such that ```answer[i]``` is guaranteed to fit in a 32-bit integer.
 

**Follow up:** Can you solve the problem in `O(1)` extra space complexity? (The output array does not count as extra space for space complexity analysis.)

````python
class Solution1:
    def productExceptSelf(self, nums: list[int]) -> list[int]:
        products = []
        for i in range(len(nums)):
            product = 1
            for j in range(len(nums)):
                if i != j:
                    product *= nums[j]
            products.append(product)

        return products
````

````python
class Solution2:
    def productExceptSelf(self, nums: list[int]) -> list[int]:
        ans = [1] * len(nums)
        prefix = 1

        # Step 1: Calculate Prefix products
        # ans[i] contains the product of all elements to the left of i
        for i in range(len(nums)):
            ans[i] *= prefix
            prefix *= nums[i]

        # Step 2: Calculate Suffix products on the fly
        # Multiply current prefix by the product of all elements to the right
        suffix = 1
        for i in range(len(nums) - 1, -1, -1):
            ans[i] *= suffix
            suffix *= nums[i]

        return ans


soln = Solution2()
result = soln.productExceptSelf([1, 2, 3, 4])
print(result)
````
