```python
class Solution:
    def findMedianSortedArrays(self, nums1: list[int], nums2: list[int]) -> float:
        n = len(nums1)
        for num in nums2:
            nums1.append(num)
        
        n = len(nums1)
        nums1 = sorted(nums1)
        if len(nums1) % 2 == 0:
            return (nums1[n // 2] + nums1[n //2 - 1]) / 2
        else:
            return nums1[n // 2] / 2
        
soln = Solution()
soln.findMedianSortedArrays([1, 2], [3, 4])
```
